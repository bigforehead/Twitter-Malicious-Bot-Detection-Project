```python
import nltk
import numpy as np
import pandas as pd
from nltk.corpus import stopwords
```

```python
count1 = 0 
count2 = 0
count3 = 0
TF_fake = [] 
df_fake = [] 
TF_scam = [] 
df_scam = [] 
TF_spam = [] 
df_spam = [] 
dict_fake = pd.read_csv('E:\MSBA\Courses\Fall\Web Analytics\Project\Phase3\Dataset\All Tweet\FinalData\Dictionary\dict_fake.csv')
dict_scam = pd.read_csv('E:\MSBA\Courses\Fall\Web Analytics\Project\Phase3\Dataset\All Tweet\FinalData\Dictionary\dict_scam.csv')
dict_spam = pd.read_csv('E:\MSBA\Courses\Fall\Web Analytics\Project\Phase3\Dataset\All Tweet\FinalData\Dictionary\dict_spam.csv')
Tweet = pd.read_csv('E:\MSBA\Courses\Fall\Web Analytics\Project\Phase3\Dataset\All Tweet\FinalData\Tweet.csv', index_col=[0])
for i in range(0, len(Tweet)): 
    tweet = nltk.FreqDist(Tweet.iloc[i, Tweet.columns.get_loc('text')].replace('b\'RT', '').replace('b"RT', '').replace('b\'', '').replace('b"', '').split(' '))
    
    for term in tweet.keys():
        if term in np.array(dict_fake['keyword']).tolist() or term.lower().startswith('http'):
            count1 += tweet[term]
    
        if term in np.array(dict_scam['keyword']).tolist() or term.lower().startswith('http'):
            count2 += tweet[term]
            
        if term in np.array(dict_spam['keyword']).tolist() or term.lower().startswith('http'):
            count3 += tweet[term]
            
    if count1 > 0:
        df_fake.append(1)
    else:
        df_fake.append(0)
    TF_fake.append(count1/len(tweet))
    
    if count2 > 0:
        df_scam.append(1)
    else:
        df_scam.append(0)
    TF_scam.append(count2/len(tweet))
    
    if count3 > 0:
        df_spam.append(1)
    else:
        df_spam.append(0)
    TF_spam.append(count3/len(tweet))
    
    count1 = 0
    count2 = 0
    count3 = 0
    print(i)
Tweet['TF_fake']=TF_fake
Tweet['df_fake']=df_fake
Tweet['TF_scam']=TF_scam
Tweet['df_scam']=df_scam
Tweet['TF_spam']=TF_spam
Tweet['df_spam']=df_spam
print(Tweet.tail(30))
```
