```python
#Crawl 250 tweets each ID
def get_all_tweets(user_id):
    alltweets = []
    
    #250 tweets for each crawling
    new_tweets = api.user_timeline(user_id=user_id, count=50)
    
    # save most recent tweets
    alltweets.extend(new_tweets)
    
    # save the id of the oldest tweet less one
    oldest = alltweets[-1].id - 1
    
    # keep grabbing tweets until there are no tweets left to grab
    while len(new_tweets) >0 and len(alltweets) <= 200:
        print(f'getting tweets before {oldest}')

        # all subsequent requests use the max_id param to prevent duplicates
        new_tweets = api.user_timeline(user_id=user_id, count=50, max_id=oldest)

        # save most recent tweets
        alltweets.extend(new_tweets)

        # update the id of the oldest tweet less one
        oldest = alltweets[-1].id - 1
        print(f'...{len(alltweets)} tweets downloaded so far')
        
    # transform the tweepy tweets into a 2D array that will populate the csv
    outtweets = [[user_id, tweet.id_str, tweet.created_at, tweet.retweet_count, tweet.favorite_count, tweet.text.encode("utf-8")] for tweet in alltweets]
    # write the csv
    with open(r'E:\MSBA\Courses\Fall\Web Analytics\Project\Phase3\Dataset\All Tweet\tweets.csv', 'a', newline='', encoding="utf-8") as f:
        writer = csv.writer(f)
        writer.writerow(["id", "tweet_id", "created_at", "retweet_count", "favorite_count", "text"])
        writer.writerows(outtweets)
    pass
if __name__ == '__main__':
    with open(r'E:\MSBA\Courses\Fall\Web Analytics\Project\Phase3\Dataset\All Tweet\ValidID.csv', 'r', encoding="utf-8") as f:
        ID = csv.reader(f)
        for row in ID:
            # pass when something wrong with ID
            try:
                get_all_tweets(row[0])
            except tweepy.TweepError as e:
                print('Failed to run the command on that user, Skipping...')
            except IndexError as e:
                print('List index out of range, Skipping...')
                continue
```
